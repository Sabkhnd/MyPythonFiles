```python
def f(n):
    num = 0
    while (num < n ) :
        yield num
        num +=1

for i in f(10):
    print(i)
```

```python
class Computer :
    Count = 0
    def __init__(self, ram, hard, cpu):
        self.ram = ram
        self.hard = hard
        self.cpu = cpu
        Computer.Count += 1
    def value(self):
        return self.ram + self.hard + self.cpu
    
    def __del__(self):
        Computer.Count -= 1
    def rcount(self):
        return (Computer.Count)
 
    

class Laptop(Computer):
    pass
#mesle computere

class Notebook(Computer):
    def __init__(self, size):
        self.size = size
    
    
pc1 = Computer(2,3,6)
pc2 = Computer(20,31,62)
pc3 = Computer(26,3,6)



print(pc1.value())
print(pc2.value())
print(pc3.value())

Laptop1 = Laptop(5,6,6)

print(Laptop1.value())
print(pc2.rcount())
```
